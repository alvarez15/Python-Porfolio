```python
from sjvisualizer import DataHandler 
from sjvisualizer import Canvas 
from sjvisualizer import BarRace
import json
```

```python
# define constants 
FPS = 60
Duration = 1
```

```python
# loding colors 
with open("colors.json") as f:
    colors = json.load(f)
    
colors
```

```python
df = DataHandler.DataHandler(excel_file = "ind_return.xlsx", number_of_frames = FPS*Duration*60).df

df
```

```python
# creating the canvas to which we want to draw all elements 
canvas = Canvas.canvas()
```

```python
# creating bar race 
bar_race =BarRace.bar_race(canvas=canvas.canvas, df = df, colors = colors)
```

```python
# adding the bar race to the Canvas 
canvas.add_sub_plot(bar_race)
```

```python
#adding the a title 
canvas.add_title("26-Year Indices Return (Normalized to 100)")
canvas.add_sub_title("From 1997 - 2023")
```

```python
# addind a time indication 
canvas.add_time(df=df, time_indicator ="year")
```

```python
 canvas.play(fps=FPS)
```
